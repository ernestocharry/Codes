# Ejemplo de separación de imagenes utilizando información de color

En este notebook se muestra un ejemplo sencillo de separación de una imagen, 
en el cual se va a utilizar la información del color de la esfera para separar 
esta del resto de objetos que se encuentran en la imagen.

El primer paso, es abrir la imagen a separar y convertirla en una matriz de NUMPY (Para que este notebook funcione, las imagenes deben estar en la misma carpeta del NB).

```python
im=imread("scene05771.png")
```

```python
imshow(im)
```

Como cada color está representado como una tripleta de numeros $(R,G,B)$, se  va a suponer que esta no representa un color, sino un punto en un espacio cartesiano con coordenadas dadas por los valores de color presentes. En este espacio cartesiano, la esfera tiene un color que está dado por la coordenada $C_e=(0.9,0.5,0.45)$ (estos valores se sacaron usando el GIMP de una imagen arbitraria. Es posible que no en todas las imagenes los valores sean los mismos, por lo que hay que ajustar estos numeros para que la separacion funcione en todas las imagenes).

El algoritmo de separacion que se propone, calcula la distancia cartesiana de cada uno de los pixeles que componen una imagen con el valor de color promedio $C_e$, y luego asume que los pixéles que pertenecen a la esfera son aquellos cuya **"distancia"** al color promedio, es pequeña.

Este procedimiento, es una simplificación de el algoritmo de segmentación kNN - (K- nearest neighbor) 

```python
## Definición de Ce

Ce=array((.9,.5,.45))

## Calculo de la diferencia entre Ce y cada uno de los pixeles de la imagen
s=im-Ce

## Calculo del modulo de las diferencias para cada uno de los pixeles
d=sqrt((s**2).sum(axis=2))


imshow(d);colorbar();title("Mapa de distancias a Ce")
```

```python
##Aplicación de un umbral para separar los pixeles que
##Pertenecen a la esfera de los que no

imshow(where(d<.2,1,0));colorbar()
```
